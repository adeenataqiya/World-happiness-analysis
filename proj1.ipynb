```python
df.sort_values(by="Score",ascending=False)
df.head(5)
```

# Visualization 1: Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(df["Score"],kde=True)
plt.title("Distribution of Happiness Scores")
plt.show()
```

# 2-Boxplot

```python
sns.boxplot(x=df["Score"])
plt.title("Boxplot of Happiness Scores")
plt.show()
```

# 3-GDP vs Happiness

```python
sns.scatterplot(data=df,x="GDP per capita",y="Score")
plt.title("GDP vs Happiness")
plt.show()
```

countries with higher GDP have high happiness

# 4-top 5 Happiest Countries

```python
top5 = df.nlargest(5, "Score")
plt.figure(figsize=(6,6))
plt.pie( top5["Score"], labels=top5["Country or region"], autopct="%1.1f%%")
plt.title("Top 5 Happiness Scores")
plt.show()
```

# 5-Top 10 Happiest Countries

```python
top10 = df.nlargest(10, "Score")
plt.figure(figsize=(10,5))
sns.lineplot(data=top10,x="Country or region",y="Score",marker="*")
plt.xticks(rotation=90)
plt.title("Top 10 Happiness Scores")
plt.show()
```

# 6-Correlation Heatmap

```python
corr = df.corr(numeric_only=True)
sns.heatmap( corr,annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
